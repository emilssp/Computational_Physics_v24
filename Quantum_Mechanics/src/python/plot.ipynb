```python
import matplotlib.pyplot as plt
import numpy as np

# Load the eigenvalues and eigenvectors
print
eigenvalues = np.loadtxt('../../data/raw/eigenvalues.txt')
eigenvectors = np.loadtxt('../../data/raw/eigenvectors.txt')

# Determine the number of points and create a position array
N = eigenvectors.shape[0]
x = np.linspace(0, 1, N+2)  # +2 to include the boundaries

# Plot the first few eigenstates
plt.figure(figsize=(10, 8))
for i in range(min(4, N)):  # Plot the first 4 eigenstates, or fewer if N<4
    plt.plot(x, np.insert(np.append(eigenvectors[:, i], 0), 0, 0), label=f'n={i+1}')

# Adding the boundaries (always zero)
plt.plot([0, 1], [0, 0], color='k', linewidth=2)

# Label the axes
plt.xlabel('Position $x$')
plt.ylabel('Wave function $\psi(x)$')

# Add a legend
plt.legend()

# Show the plot
plt.title('Wave Functions for the Infinite Potential Well')
plt.show()
```
